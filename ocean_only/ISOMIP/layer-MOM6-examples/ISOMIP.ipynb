```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

```python
prog = xr.open_dataset('prog__0001_001.nc')
IClayerpres = xr.open_dataset('archive/ISOMIP_IC-pressurelayer.nc')
IClayernopres = xr.open_dataset('archive/ISOMIP_IC-nopressurelayer.nc')
ICsigmazstarnopres = xr.open_dataset('archive/ISOMIP_IC-nopressuresigmazstar.nc')
ICtest = xr.open_dataset('ISOMIP_IC.nc')
pyalesigmazstar = xr.open_dataset('INPUT/sigmazstar.nc').h
```

```python
pyalesigmazstar.sum('nz').plot()
```

```python
fig, axes = plt.subplots(dpi = 200)
plt.plot(np.arange(240),IClayerpres.h.isel(Time = 0).sum('Layer').isel(lath = 20).values, label = 'layer IC, surf pres')
plt.plot(np.arange(240),IClayernopres.h.isel(Time = 0).sum('Layer').isel(lath = 20).values, label = 'layer IC, no surf pres')
plt.plot(np.arange(240),ICsigmazstarnopres.h.isel(Time = 0).sum('Layer').isel(lath = 20).values, label = 'sigmazstar IC, no surf pres')
plt.plot(np.arange(240),pyalesigmazstar.sum('nz').isel(ny = 20).values, label = 'py-ale sigmashelfzstar')

plt.plot(np.arange(240),ICtest.h.isel(Time = 0).sum('Layer').isel(lath = 20).values, label = 'IC test')

plt.legend()
plt.title(r'$\Sigma_z h$')
```

```python
fig, axes = plt.subplots(dpi = 200)
plt.plot(np.arange(240),(IClayerpres.h-IClayernopres.h).isel(Time = 0).sum('Layer').isel(lath = 20).values, 
         label = 'layer IC, surf pres - no surf pres')
plt.legend()
plt.title(r'$\Sigma_z h$')
```

```python
fig, axes = plt.subplots(dpi = 200)
plt.plot(np.arange(240),(IClayernopres.h).isel(Time = 0).sum('Layer').isel(lath = 20).values -
         ICsigmazstarnopres.h.isel(Time = 0).sum('Layer').isel(lath = 20), 
         label = 'layer IC, no surf pres - sigmazstarno surf pres')
plt.legend()
plt.title(r'$\Sigma_z h$')
```

```python
fig, axes = plt.subplots(dpi = 200)

plt.plot(np.arange(240),pyalesigmazstar.sum('nz').isel(ny = 20).values-
         ICtest.h.isel(Time = 0).sum('Layer').isel(lath = 20).values, label = 'py-ale sigmashelfzstar - IC test')

plt.legend()
plt.title(r'$\Sigma_z h$')
```

```python
fig, axes = plt.subplots(dpi = 200)

plt.plot(np.arange(240),prog.h.sum('zl').isel(yh = 20,Time = -1).values-
         ICtest.h.isel(Time = 0).sum('Layer').isel(lath = 20).values, label = 'time = 5days - IC test')

plt.legend()
plt.title(r'$\Sigma_z h$')
```
